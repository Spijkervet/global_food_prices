```python
import pandas as pd
import numpy as np

from bokeh.plotting import figure
from bokeh.io import output_file, show
from matplotlib import pyplot as plt

%matplotlib inline

# get database as panda dataframe
df = pd.read_csv('WFPVAM_FoodPrices_version2_Retail.csv')

COUNTRY = 'adm0_name'
REGION = 'adm1_name'
CITY = 'mkt_name'
PROD = 'cm_name'
CURR = 'cur_name'
SELLER = 'pt_name'
UNIT = 'um_name'
MONTH = 'mp_month'
YEAR = 'mp_year'
DATE = 'date'
PRICE = 'mp_price'
SOURCE = 'mp_commoditysource'
```

```python
#print(df[PROD].unique())
is_bread = df[PROD] == "Bread"
print(df[is_bread][PRICE].value_counts())
x = df[is_bread][DATE]

x = df[DATE]
y = df[PRICE]

# Prepare the output file
output_file("Line_from_csv.html")

# Create a figure object
f = figure()

# Create line plot
f.line(x,y)

show(f)
```
